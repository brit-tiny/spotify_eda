#### Duplicated date
##### Returns a boolean
- True = duplicate data
- False = data is not duplicated
##### Keep parameter
- first: the first time the data is retuned it will be false, any subsequent occurance will return true
- last: the last time the data is retuned it will be false, any prior occurance will return true
- false: any duplicated data will return true


```python
df. duplicated(keep=False).sum
```

```python
df.loc[df.duplicated(keep=False)].sort_values('track_id')
```

#### This line of code points to duplicated data withing in 'track_id" showing the only difference is the genre 
##### This can:
- influence how the data is viewed later down the line
- help us decide what data to keep

```python
df.loc[df['track_id'].duplicated(keep=False)].sort_values('track_id')
```
